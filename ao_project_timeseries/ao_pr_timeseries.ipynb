## –û–±—É—á–µ–Ω–∏–µ

**–ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤**

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –ø—Ä–∏ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏–∏
def get_features(df_learning):
    # –î–æ–±–∞–≤–ª—è–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
    df_learning['month'] = df_learning.index.month  
    df_learning['day'] = df_learning.index.day  
    df_learning['dayofweek'] = df_learning.index.dayofweek  
    df_learning['hour'] = df_learning.index.hour 

    # –î–æ–±–∞–≤–ª—è–µ–º –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –ª–∞–≥–∏ –¥–ª—è —É—á—ë—Ç–∞ –ø—Ä–æ—à–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∑–∞–∫–∞–∑–æ–≤
    for lag in range(1, 6):  
        df_learning[f'lag_{lag}'] = df_learning['num_orders'].shift(lag)

    # –î–æ–±–∞–≤–ª—è–µ–º —Å–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ –¥–ª—è —Å–≥–ª–∞–∂–∏–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö
    df_learning['rolling_mean'] = df_learning['num_orders'].shift().rolling(1).mean()

    # –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, –ø–æ—è–≤–∏–≤—à–∏–º–∏—Å—è –ø–æ—Å–ª–µ —Å–æ–∑–¥–∞–Ω–∏—è –ª–∞–≥–æ–≤
    df_learning.dropna(inplace=True)

    return df_learning
```

–ü—Ä–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, —á—Ç–æ–±—ã –æ–Ω–∞ –≤–æ–∑–≤—Ä–∞—â–∞–ª–∞ –Ω–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –∞ –Ω–µ –∏–∑–º–µ–Ω—è–ª–∞ –≥–ª–æ–±–∞–ª—å–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é.

–°—Ç–æ–∏—Ç –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –∑–∞–¥–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä max_lag —Ä–∞–≤–Ω—ã–º 24 –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Ñ—É–Ω–∫—Ü–∏–µ–π. –í —Å–ª—É—á–∞–µ –∑–∞–∫–∞–∑–æ–≤ —Ç–∞–∫—Å–∏ —Å–∞–º–∞—è –æ—á–µ–≤–∏–¥–Ω–∞—è —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å - –≤–Ω—É—Ç—Ä–∏ —Å—É—Ç–æ–∫, –∞ –≤ —Å—É—Ç–∫–∞—Ö 24 —á–∞—Å–∞, —ç—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –±—É–¥–µ—Ç –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º. –†–∞–∑–º–µ—Ä –æ–∫–Ω–∞ —Ç–æ–∂–µ –º–æ–∂–Ω–æ —É–≤–µ–ª–∏—á–∏—Ç—å

```python
# –ö–æ–ø–∏—Ä—É–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–µ—Ä–µ–¥ –æ–±—Ä–∞–±–æ—Ç–∫–æ–π
df_learning = df.copy()
df_learning = get_features(df_learning)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞–∑–º–µ—Ä—ã –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –∏ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
print("–†–∞–∑–º–µ—Ä –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞:", df.shape)
print("–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞ –ø–æ—Å–ª–µ —Å–æ–∑–¥–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:", df_learning.shape)
```

```python
# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ (features) –∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ (target)

features = df_learning.drop(['num_orders'], axis=1)  
target = df_learning['num_orders']
```

```python
# –†–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
features_train, features_test, target_train, target_test = train_test_split(
    features, target, shuffle=False, test_size=0.1, random_state=17)

# –í—ã–≤–µ–¥–µ–º —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫
print("–†–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏:", features_train.shape, target_train.shape)
print("–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏:", features_test.shape, target_test.shape)
```

**–û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π**

```python
# –°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π –∏ –∏—Ö –º–µ—Ç—Ä–∏–∫
train_models = []
```

–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

```python
# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model = LinearRegression()

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
model.fit(features_train, target_train)

# –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
predictions = model.predict(features_train)

# –í—ã—á–∏—Å–ª—è–µ–º RMSE 
rmse = mean_squared_error(target_train, predictions, squared=False)

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(f'RMSE –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ: {rmse:.2f}')

# –î–æ–±–∞–≤–ª—è–µ–º –º–æ–¥–µ–ª—å –∏ –µ—ë RMSE –≤ —Å–ø–∏—Å–æ–∫ –æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π
train_models.append(["Linear Regression", model, rmse])
```

–î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

```python
# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π
model = DecisionTreeRegressor(random_state=12345)

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
parameters = {'max_depth': range(1, 11, 2)}

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤ (—Ä–∞–∑–±–∏–≤–∞–µ–º –Ω–∞ 2 —á–∞—Å—Ç–∏)
tscv = TimeSeriesSplit(n_splits=2)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º GridSearchCV –¥–ª—è –ø–æ–¥–±–æ—Ä–∞ –ª—É—á—à–µ–π –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞
gsearch = GridSearchCV(estimator=model, cv=tscv, param_grid=parameters)

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
gsearch.fit(features_train, target_train)

# –í—ã–≤–æ–¥–∏–º –Ω–∞–∏–ª—É—á—à–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä max_depth
best_params = gsearch.best_params_
print(f'–õ—É—á—à–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä –¥–ª—è Decision Tree: {best_params}')

# –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ —Å –ª—É—á—à–µ–π –º–æ–¥–µ–ª—å—é
predictions = gsearch.predict(features_train)

# –í—ã—á–∏—Å–ª—è–µ–º RMSE 
rmse = mean_squared_error(target_train, predictions, squared=False)

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(f'RMSE Decision Tree –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ: {rmse:.2f}')

# –î–æ–±–∞–≤–ª—è–µ–º –º–æ–¥–µ–ª—å –∏ –µ—ë RMSE –≤ —Å–ø–∏—Å–æ–∫ –æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π
train_models.append(["Decision Tree", gsearch.best_estimator_, rmse])
```

–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

```python
# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
model = RandomForestRegressor(random_state=12345)

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
parameters = {
    'n_estimators': range(50, 100, 10),  
    'max_depth': range(1, 11, 2)  
}

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤ (—Ä–∞–∑–±–∏–≤–∞–µ–º –Ω–∞ 2 —á–∞—Å—Ç–∏)
tscv = TimeSeriesSplit(n_splits=2)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º GridSearchCV –¥–ª—è –ø–æ–¥–±–æ—Ä–∞ –ª—É—á—à–∏—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
gsearch = GridSearchCV(estimator=model, cv=tscv, param_grid=parameters, n_jobs=-1)

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å —Å –ø–æ–¥–±–æ—Ä–æ–º –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
gsearch.fit(features_train, target_train)

# –í—ã–≤–æ–¥–∏–º –Ω–∞–∏–ª—É—á—à–∏–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
best_params = gsearch.best_params_
print(f'–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è Random Forest: {best_params}')

# –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ —Å –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π –º–æ–¥–µ–ª—å—é
predictions = gsearch.predict(features_train)

# –í—ã—á–∏—Å–ª—è–µ–º RMSE 
rmse = mean_squared_error(target_train, predictions, squared=False)

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç RMSE
print(f'RMSE Random Forest –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ: {rmse:.2f}')

# –î–æ–±–∞–≤–ª—è–µ–º –º–æ–¥–µ–ª—å –∏ –µ—ë RMSE –≤ —Å–ø–∏—Å–æ–∫ –æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π
train_models.append(["Random Forest", gsearch.best_estimator_, rmse])
```

## –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ

```python
# –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö

# –°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
test_models = []

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –º–æ–¥–µ–ª–∏ –∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è RMSE
def display_result(target, pred, model_name, rmse):
    """
    –§—É–Ω–∫—Ü–∏—è —Å—Ç—Ä–æ–∏—Ç –∫—Ä–∞—Å–∏–≤—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π,
    –∞ —Ç–∞–∫–∂–µ –≤—ã–≤–æ–¥–∏—Ç RMSE.
    """
    result = target.to_frame().reset_index() 
    result['prediction'] = pd.Series(pred) 
    result.set_index('datetime', inplace=True)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è 
    plt.figure(figsize=(15, 5))
    sns.set_style("whitegrid") 
    sns.set_palette("pastel")
    plt.plot(result.index, result['num_orders'], label='–§–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è', color=sns.color_palette("pastel")[0], linewidth=2)
    plt.plot(result.index, result['prediction'], label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è', color=sns.color_palette("pastel")[3], linestyle='dashed', linewidth=2)

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–µ–π
    plt.title(f'{model_name} - RMSE={rmse:.2f}', fontsize=14, fontweight='bold', color='darkblue')
    plt.xlabel('–î–∞—Ç–∞', fontsize=12, color='black')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤', fontsize=12, color='black')
    plt.xticks(rotation=25)
    
    # –î–æ–±–∞–≤–ª—è–µ–º –ª–µ–≥–µ–Ω–¥—É –∏ —Å–µ—Ç–∫—É
    plt.legend(fontsize=12)
    plt.grid(alpha=0.3)

    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
    plt.show()

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –≤—Å–µ –æ–±—É—á–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
for model_name, trained_model, _ in train_models:
    predictions = trained_model.predict(features_test)  # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏
    rmse = mean_squared_error(target_test, predictions, squared=False)  # –í—ã—á–∏—Å–ª—è–µ–º RMSE

    # –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    display_result(target_test, predictions, model_name, rmse)

    # –î–æ–±–∞–≤–ª—è–µ–º RMSE –≤ —Å–ø–∏—Å–æ–∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
    test_models.append([model_name, rmse])
```
