```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (pie-charts)

# –£–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã –¥–æ–ª–µ–π
def plot_pie(df: pd.DataFrame, column: str, title: str):
    """
    –°—Ç—Ä–æ–∏—Ç –∫—Ä—É–≥–æ–≤—É—é –ø–æ–Ω—á–∏–∫–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Å –Ω–æ–≤–æ–π —Ü–≤–µ—Ç–æ–≤–æ–π –ø–∞–ª–∏—Ç—Ä–æ–π.
    """
    fig, ax = plt.subplots(figsize=(6, 6))
    
    
    colors = sns.color_palette("Set3")
    
    (
        df[column]
        .value_counts(normalize=True)
        .plot.pie(
            ax=ax,
            colors=colors,
            autopct="%1.0f%%",
            startangle=90,
            wedgeprops={"linewidth": 1, "edgecolor": "white"},
            pctdistance=0.85
        )
    )
    
    ax.set_ylabel("")
    ax.set_title(title, pad=15)
    
    # –ü–æ–Ω—á–∏–∫–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
    centre_circle = plt.Circle((0, 0), 0.70, fc="white")
    fig.gca().add_artist(centre_circle)
    
    plt.tight_layout()
    plt.show()
```

```python
# –î–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plot_pie(contract_new, "Type", "–¢–∏–ø –æ–ø–ª–∞—Ç—ã")
plot_pie(contract_new, "PaperlessBilling","–ù–∞–ª–∏—á–∏–µ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –∫–≤–∏—Ç–∞–Ω—Ü–∏–∏")
plot_pie(contract_new, "PaymentMethod",   "–°–ø–æ—Å–æ–± –æ–ø–ª–∞—Ç—ã")
```

 <div class="alert alert-success">
<b>üëç –£—Å–ø–µ—Ö:</b> –í—Å–µ –≤–µ—Ä–Ω–æ!

```python
def f_countplot(data, suptitle, xlabel, ylabel):
    fig, ax1 = plt.subplots(1, 1, figsize=(15, 15))
    fig.tight_layout(h_pad=3)
    fig.suptitle(suptitle, x=0.5, y=1.01, fontsize=16)
    sns.countplot(y=data, ax=ax1)
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)
```

```python
# –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è BeginDate
f_countplot(
    data=sorted(contract_new['BeginDate']),
    suptitle='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–ª—é—á—ë–Ω–Ω—ã—Ö –¥–æ–≥–æ–≤–æ—Ä–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º',
    xlabel='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–æ–≥–æ–≤–æ—Ä–æ–≤',
    ylabel='–î–∞—Ç–∞ –Ω–∞—á–∞–ª–∞'
)

# –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è EndDate
f_countplot(
    data=sorted(contract_new[contract_new['EndDate'] != 'No']['EndDate']),
    suptitle='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞—Å—Ç–æ—Ä–≥–Ω—É—Ç—ã—Ö –¥–æ–≥–æ–≤–æ—Ä–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º',
    xlabel='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–æ–≥–æ–≤–æ—Ä–æ–≤',
    ylabel='–î–∞—Ç–∞ —Ä–∞—Å—Ç–æ—Ä–∂–µ–Ω–∏—è'
)
```

</div>

```python
# –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ —Å —É—á—ë—Ç–æ–º —Ç–∏–ø–∞ –ø–∞–∫–µ—Ç–∞ —É—Å–ª—É–≥
# –°–ø–∏—Å–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –∑–∞–º–µ–Ω—ã
internet_cols  = ['InternetService', 'OnlineSecurity', 'OnlineBackup',
                  'DeviceProtection', 'TechSupport', 'StreamingTv', 'StreamingMovies']
phone_cols     = ['MultipleLines']

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –º–∞—Ä–∫–µ—Ä–∞–º–∏ "I" (–Ω–µ—Ç –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–ø–∞–∫–µ—Ç–∞) –∏ "T" (–Ω–µ—Ç —Ç–µ–ª–µ—Ñ–æ–Ω–Ω–æ–≥–æ –ø–∞–∫–µ—Ç–∞)
df[phone_cols]    = df[phone_cols].fillna('T')
df[internet_cols] = df[internet_cols].fillna('I')

# –ö–æ–Ω—Ç—Ä–æ–ª—å: –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –ª–∏ –ø—Ä–æ–ø—É—Å–∫–∏?
display(df[phone_cols + internet_cols].isna().sum())
```

–ü—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å; –º–∞—Ä–∫–µ—Ä—ã "I" –∏ "T" –ø–æ–∑–≤–æ–ª—è—Ç –º–æ–¥–µ–ª–∏ –æ—Ç–ª–∏—á–∞—Ç—å –∫–ª–∏–µ–Ω—Ç–æ–≤, –Ω–µ –ø–æ–¥–∫–ª—é—á–∞–≤—à–∏—Ö —É—Å–ª—É–≥—É –≤–æ–≤—Å–µ, –æ—Ç —Ç–µ—Ö, –∫—Ç–æ –æ—Ç–∫–ª—é—á–∏–ª –µ—ë –ø–æ–∑–∂–µ.

## –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ (phi-k) –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞

```python
# –§–∏–∫—Å–∏—Ä—É–µ–º CustomerId –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–∞ –∏ —É–±–∏—Ä–∞–µ–º –¥–∞—Ç—ã
df = df.set_index("CustomerId").drop(columns=["BeginDate", "EndDate"])

# –ß–∏—Å–ª–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ —Ç—Ä–∞–∫—Ç–æ–≤–∞—Ç—å –∫–∞–∫ –∏–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω—ã–µ –¥–ª—è phi-k
interval_cols = ["MonthlyCharges", "TotalCharges", "DurationContract"]

# –°—Ç—Ä–æ–∏–º –º–∞—Ç—Ä–∏—Ü—É phi-k –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º
plt.figure(figsize=(15, 15))
plt.title("–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π phi(k)", fontsize=16, pad=12)

corr_matrix = df.phik_matrix(interval_cols=interval_cols).round(2)
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar_kws={"label": "phi-k"},
    linewidths=0.5
)

plt.show()
```

## –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∫ –æ–±—É—á–µ–Ω–∏—é –º–æ–¥–µ–ª–µ–π

```python
# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤

# –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∏ —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
cat_col_names = ["Type", "PaymentMethod", "Partner", "OnlineBackup", "MultipleLines"]
num_col_names = ["MonthlyCharges", "TotalCharges", "DurationContract"]

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ —Ç–∏–ø 'category'
df_reduced[cat_col_names] = df_reduced[cat_col_names].astype("category")

# –ö–æ–Ω—Ç—Ä–æ–ª—å
df_reduced.info()
```

```python
# –§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫
# –°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ train : test = 3 : 1, random_state = 210425.

X = df_reduced.drop(columns="Target")
y = df_reduced["Target"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
)

print(
    "–†–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫:",
    f"train = {X_train.shape[0]} ({X_train.shape[0]/X.shape[0]:.2%}) |",
    f"test = {X_test.shape[0]} ({X_test.shape[0]/X.shape[0]:.2%})"
)
```

```python
# –ü–∞–π–ø–ª–∞–π–Ω –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏

ohe_pipe = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(drop="first", handle_unknown="ignore", sparse_output=False))
    ]
)

data_preprocessor = ColumnTransformer(
    transformers=[
        ("cat", ohe_pipe, cat_col_names),
        ("num", StandardScaler(), num_col_names)
    ],
    remainder="passthrough"
)
```

<b>ü§î –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è:</b> SimpleImputer –∑–¥–µ—Å—å –Ω–µ –Ω—É–∂–µ–Ω
</div>

### –í—ã–≤–æ–¥ –ø–æ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–µ –¥–∞–Ω–Ω—ã—Ö


1. **–ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ —Ç–∏–ø–æ–≤**  
   - –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ (`Type`, `PaymentMethod`, `Partner`, `OnlineBackup`, `MultipleLines`) –ø–µ—Ä–µ–≤–µ–¥–µ–Ω—ã –≤ —Ç–∏–ø `category`, —á—Ç–æ –æ–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ—Ç –ø–∞–º—è—Ç—å –∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç—Å—è `OneHotEncoder`.  
   - –ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ (`MonthlyCharges`, `TotalCharges`, `DurationContract`) –æ—Å—Ç–∞–≤–ª–µ–Ω—ã –≤ —Ñ–æ—Ä–º–∞—Ç–µ `float64` –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–µ–π —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏.


2. **–§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫**  
   - –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –≤ —Ç—Ä–µ–±—É–µ–º–æ–º —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ **train : test = 3 : 1** (`test_size = 0.25`, `random_state = 210425`, —Å—Ç—Ä–∞—Ç–∏—Ñ–∏–∫–∞—Ü–∏—è –ø–æ —Ü–µ–ª–µ–≤–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É).  
   - –ò—Ç–æ–≥–æ–≤—ã–µ —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –∑–∞–¥–∞–Ω–∏—é –∏ –æ–±–µ—Å–ø–µ—á–∏–≤–∞—é—Ç —Ä–µ–ø—Ä–µ–∑–µ–Ω—Ç–∞—Ç–∏–≤–Ω–æ—Å—Ç—å.


–î–∞–Ω–Ω—ã–µ –ø–æ–ª–Ω–æ—Å—Ç—å—é –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω—ã –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —ç—Ç–∞–ø—É - **–æ–±—É—á–µ–Ω–∏—é –º–æ–¥–µ–ª–µ–π –∏ –ø–æ–¥–±–æ—Ä—É –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤**.

## –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π –∏ –ø–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤

**–í—ã–±–æ—Ä –º–µ—Ç—Ä–∏–∫–∏: ROC-AUC**


–î–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –±—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –º–µ—Ç—Ä–∏–∫–∞ **ROC-AUC**, —Å —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –ø–æ—Ä–æ–≥–æ–º **‚â• 0.85** (–≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º–∏ –ø—Ä–æ–µ–∫—Ç–∞).


*–°–ø—Ä–∞–≤–∫–∞ –ø–æ –º–µ—Ç—Ä–∏–∫–µ:*


- **ROC-–∫—Ä–∏–≤–∞—è (Receiver Operating Characteristic Curve)** - –≥—Ä–∞—Ñ–∏–∫, –æ—Ç–æ–±—Ä–∞–∂–∞—é—â–∏–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É:
  - **True Positive Rate (TPR)** - –¥–æ–ª–µ–π –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤,
  - **False Positive Rate (FPR)** - –¥–æ–ª–µ–π –ª–æ–∂–Ω–æ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–≥–Ω–æ–∑–æ–≤, –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö –ø–æ—Ä–æ–≥–∞—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.


- **AUC (Area Under Curve)** - –ø–ª–æ—â–∞–¥—å –ø–æ–¥ ROC-–∫—Ä–∏–≤–æ–π:
  - –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –æ—Ç **0.5** (–º–æ–¥–µ–ª—å –Ω–µ –ª—É—á—à–µ —Å–ª—É—á–∞–π–Ω–æ–π) –¥–æ **1.0** (–∏–¥–µ–∞–ª—å–Ω–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤),
  - —á–µ–º –±–ª–∏–∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ AUC –∫ 1.0, —Ç–µ–º **–ª—É—á—à–µ –º–æ–¥–µ–ª—å —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å —Ä–∞–∑–ª–∏—á–µ–Ω–∏–µ–º –∫–ª–∞—Å—Å–æ–≤**.


> **–ü–æ—Ä–æ–≥ –æ—Ç—Å–µ—á–∫–∏ –ø–æ –∑–∞–¥–∞–Ω–∏—é: ROC-AUC ‚â• 0.85**

*–ú–æ–¥–µ–ª—å 1 - **RandomForestClassifier** —Å RandomizedSearchCV*

```python
# –§–∏–Ω–∞–ª—å–Ω—ã–π –∫–æ–Ω–≤–µ–π–µ—Ä: –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ + –º–æ–¥–µ–ª—å
pipe_rf = Pipeline(
    steps=[
        ("prep", data_preprocessor),
        ("rf",  RandomForestClassifier(random_state=RANDOM_STATE))
    ]
)

# –°–µ—Ç–∫–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
param_rf = {
    "rf__n_estimators": np.random.randint(80, 401, 15),
    "rf__max_depth": np.random.randint(4, 31, 10),
    "rf__min_samples_leaf": np.random.randint(1, 11, 10),
    "rf__max_features": ["sqrt", "log2", None],
    "rf__bootstrap": [True, False],
    "rf__class_weight": [None, "balanced"],
    "prep__num": [StandardScaler(), MinMaxScaler(), RobustScaler(), "passthrough"]
}
```

```python
# RandomizedSearchCV —Å –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π 5-fold
rf_search = RandomizedSearchCV(
    estimator=pipe_rf,
    param_distributions=param_rf,
    n_iter=30,
    scoring="roc_auc",
    cv=5,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    verbose=1
)
```

```python
%%time
rf_search.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã RandomForest:\n", rf_search.best_params_)
print(f"ROC-AUC (CV) = {rf_search.best_score_:.3f}")
```

*–ú–æ–¥–µ–ª—å 2 - **LightGBM** (LGBMClassifier) —Å RandomizedSearchCV*

```python
# –§–∏–Ω–∞–ª—å–Ω—ã–π –∫–æ–Ω–≤–µ–π–µ—Ä: –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ + –º–æ–¥–µ–ª—å LGBM
pipe_lgbm = Pipeline(
    steps=[
        ("prep", data_preprocessor),
        ("lgb",  LGBMClassifier(random_state=RANDOM_STATE))
    ]
)

# –°–µ—Ç–∫–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è LightGBM
param_lgbm = {
    "lgb__learning_rate": [0.005, 0.01, 0.05, 0.1],
    "lgb__n_estimators": [100, 200, 300],
    "lgb__num_leaves": [31, 63, 127],
    "lgb__min_child_samples": [10, 20, 30, 40],
    "lgb__class_weight":  [None, "balanced"],
    "prep__num": [StandardScaler()]
}
```

```python
# RandomizedSearchCV —Å 5-–∫—Ä–∞—Ç–Ω–æ–π –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π –∏ –º–µ—Ç—Ä–∏–∫–æ–π ROC-AUC
lgbm_search = RandomizedSearchCV(
    estimator=pipe_lgbm,
    param_distributions=param_lgbm,
    n_iter=30,
    scoring="roc_auc",
    cv=5,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    verbose=1
)
```

```python
%%time
lgbm_search.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã LightGBM:\n", lgbm_search.best_params_)
print(f"ROC-AUC (CV) = {lgbm_search.best_score_:.3f}")
```

*–ú–æ–¥–µ–ª—å 3 - **CatBoostClassifier** —Å RandomizedSearchCV*

```python
# –ü–µ—Ä–µ–¥–∞—ë–º —Å–ø–∏—Å–æ–∫ –∏–º—ë–Ω –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
cbc_model = CatBoostClassifier(
    random_state=RANDOM_STATE,
    cat_features=cat_col_names,
    verbose=0,
    loss_function="Logloss"
)
```

```python
# –°–µ—Ç–∫–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞
param_cbc = {
    "learning_rate": 10 ** np.random.uniform(-3, -1, 20),
    "depth": np.random.randint(4, 11, 20),
    "n_estimators": np.random.randint(128, 401, 20),
    "l2_leaf_reg": np.random.uniform(1, 8, 20),
    "border_count": np.random.choice([32, 64, 128], 20),
    "scale_pos_weight": np.random.uniform(0.5, 3.0, 20)
}
```

```python
cbc_search = RandomizedSearchCV(
    estimator=cbc_model,
    param_distributions=param_cbc,
    n_iter=30,
    scoring="roc_auc",
    cv=5,
    n_jobs=-1,
    random_state=RANDOM_STATE,
    verbose=1,
    error_score="raise"
)
```

```python
%%time
cbc_search.fit(X_train, y_train)

print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã CatBoost:\n", cbc_search.best_params_)
print(f"ROC-AUC (CV) = {cbc_search.best_score_:.3f}")
```

```python
# –°–±–æ—Ä —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏
results = pd.DataFrame({
    "Model": ["RandomForestClassifier", "LGBMClassifier", "CatBoostClassifier"],
    "ROC_AUC_Valid": [rf_search.best_score_, lgbm_search.best_score_, cbc_search.best_score_],
    "Code": ["rf_search", "lgbm_search", "cbc_search"]
})

display(results)
```

```python
# –û—Ç–±–æ—Ä –º–æ–¥–µ–ª–µ–π, –ø—Ä–µ–≤—ã—à–∞—é—â–∏—Ö –ø–æ—Ä–æ–≥ 0.85, –∏ –≤—ã–±–æ—Ä –ª—É—á—à–µ–π
top_model = (
    results[results["ROC_AUC_Valid"] >= 0.85]
    .sort_values("ROC_AUC_Valid", ascending=False)
    .iloc[0]
)

print("–í—ã–±—Ä–∞–Ω–∞ –º–æ–¥–µ–ª—å:", top_model["Model"])
print("ROC-AUC (CV):", round(top_model["ROC_AUC_Valid"], 3))
```

*–û—Ü–µ–Ω–∫–∞ **LGBMClassifier** –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:*

```python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å, –Ω–∞–π–¥–µ–Ω–Ω—É—é –Ω–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏
best_cbc = lgbm_search.best_estimator_

# –†–∞—Å—á—ë—Ç ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
roc_auc_test = roc_auc_score(y_test, best_cbc.predict_proba(X_test)[:, 1])

print(f"ROC-AUC (–∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è): {lgbm_search.best_score_:.5f}")
print(f"ROC-AUC (—Ç–µ—Å—Ç):            {roc_auc_test:.5f}")
print(
    "–ò–∑–º–µ–Ω–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ (test vs CV): ",
    f"{((roc_auc_test - lgbm_search.best_score_) / lgbm_search.best_score_) * 100:.3f} %"
)
```

```python
# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ ROC-–∫—Ä–∏–≤–æ–π –¥–ª—è LightGBMClassifier
sns.set_theme(style="whitegrid", font_scale=1.1)

# –†–∞—Å—á—ë—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç ROC-–∫—Ä–∏–≤–æ–π
fpr, tpr, _ = roc_curve(y_test, best_cbc.predict_proba(X_test)[:, 1])

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, label=f"LGBMClassifier (AUC = {roc_auc_test:.3f})", linewidth=2)
plt.plot([0, 1], [0, 1], linestyle="--", color="grey", label="–°–ª—É—á–∞–π–Ω–∞—è –º–æ–¥–µ–ª—å (AUC = 0.5)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC-–∫—Ä–∏–≤–∞—è –¥–ª—è LGBMClassifier")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

```python
# –ö–ª—é—á–µ–≤—ã–µ –º–µ—Ç—Ä–∏–∫–∏
acc = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(f"Accuracy (–¥–æ–ª—è –≤–µ—Ä–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤): {acc:.3f}")
print(f"Recall (True Positive Rate, FN‚Äë–æ—à–∏–±–∫–∏): {recall:.3f}")
print(f"Precision (Positive Predictive Value, FP): {precision:.3f}")
```
